```python
import altair as alt
```

```python
BASE_DIR='/home/thanuja/Dropbox/Lansing/'
```

```python
'''
pip install openpyxl
pip refers to Python 2 as a default in Ubuntu, this means that pip install x will install the module for Python 2 and not for 3
pip3 refers to Python 3, it will install the module for Python 3
openpyxl is a Python library to read/write Excel 2010 xlsx/xlsm/xltx/xltm files.
'''
#import all csvs and combine into one single dataframe
import pandas as pd
lansing_2015_df = pd.read_excel(BASE_DIR + "Trash_Grass_Violations_2015.xlsx",skiprows=6)
#print(lansing_2015_df.head(10))
#print(lansing_2015_df.columns)
print("total rows in lansing_2015_df: ",len(lansing_2015_df.index))
lansing_2016_df = pd.read_excel(BASE_DIR + "Trash_Grass_Violations_2016.xlsx",skiprows=6)
#print(lansing_2016_df.head(10))
print("total rows in lansing_2016_df: ",len(lansing_2016_df.index))
lansing_2017_df = pd.read_excel(BASE_DIR + "Trash_Grass_Violations_2017.xlsx",skiprows=6)
#print(lansing_2017_df.head(10))
print("total rows in lansing_2017_df: ",len(lansing_2017_df.index))
lansing_2018_df = pd.read_excel(BASE_DIR + "Trash_Grass_Violations_2018.xlsx",skiprows=6)
#print(lansing_2018_df.head(10))
print("total rows in lansing_2018_df: ",len(lansing_2018_df.index))
lansing_2019_df = pd.read_excel(BASE_DIR + "Trash_Grass_Violations_2019.xlsx",skiprows=6)
#print(lansing_2019_df.head(10))
print("total rows in lansing_2019_df ",len(lansing_2019_df.index))

lansing_dfs = [lansing_2015_df, lansing_2016_df,lansing_2017_df,lansing_2018_df,lansing_2019_df]

#Giving common column names to all lansing dataframes
for df in lansing_dfs:
    df.columns = ['Violator Address', 'Date Filed', 'Owner Address', 'Violation']

#combined dataframe
lansing_df = pd.concat(lansing_dfs,ignore_index=True)
print("total rows in lansing_df: ",len(lansing_df.index))
print("\n")
#first 30 rows
print("*****************First 30 rows******************")
print(lansing_df.head(30))
print("\n")
# last 30 rows
print("*****************Last 30 rows******************")
lansing_df.tail(30)
```

```python
#***Violations - year breakdown***
lansing_addr_df = lansing_df[["Violator Address","Violation"]]
lansing_addr_df = lansing_addr_df.pivot_table(index='Violator Address', columns='Violation', aggfunc=len, fill_value=0).reset_index()
lansing_addr_df['total'] = lansing_addr_df['Grass/Weed'] + lansing_addr_df['Trash']
lansing_addr_df.sort_values(by=['total'], inplace=True, ascending=False)
#len(lansing_addr_df)
lansing_addr_df
```

```python
violations_gt_than_15_df =lansing_addr_df[lansing_addr_df['total'] > 15]
#there are 55 addresses with more than 15 violations
violations_gt_than_15_chart = alt.Chart(violations_gt_than_15_df).mark_bar().encode(
    x=alt.X("total:Q"),
    y=alt.Y('Violator Address:N',sort='-x'),
    tooltip=['Grass/Weed:N','Trash:N']
   
).properties(width=800,height=600)
violations_gt_than_15_chart
```

```python
lansing_geocodes_df = pd.read_csv(BASE_DIR + "lansing_geocodes.csv")
lansing_geocodes_df['Input Address']=lansing_geocodes_df['Input Address'].apply(lambda x: x.split(',')[0])
merge_latlons = lansing_addr_df.merge(lansing_geocodes_df,left_on=["Violator Address"],right_on=["Input Address"], how='inner')
merge_latlons.head(10)
violations_gt_than_15_latlon_df = merge_latlons[merge_latlons['total'] > 15]

violations_gt_than_15_latlon_df=violations_gt_than_15_latlon_df.dropna(subset=['Latitude'])
violations_gt_than_15_latlon_df=violations_gt_than_15_latlon_df.dropna(subset=['Longitude'])
violations_gt_than_15_latlon_df
#len(violations_gt_than_15_latlon_df) : 47
```

```python
import folium
violations_gt_than_15_map = folium.Map(location = [violations_gt_than_15_latlon_df['Latitude'].mean(), violations_gt_than_15_latlon_df['Longitude'].mean()], 
                   width = 1000, 
                   height = 800,
                   zoom_start = 12, 
                   tiles = 'openstreetmap')

for lat, lng in zip(violations_gt_than_15_latlon_df['Latitude'], violations_gt_than_15_latlon_df['Longitude']):    
    residence = folium.CircleMarker(
            location=[lat, lng],
            radius=8,
            color='red',
            fill=True,
            fill_color='red')
    residence.add_to(violations_gt_than_15_map)
violations_gt_than_15_map
```

```python
#***Violations - year breakdown***
lansing_df['year'] = pd.DatetimeIndex(lansing_df['Date Filed']).year
lansing_yr_df = lansing_df[["year","Violation"]]
lansing_yr_df = lansing_yr_df.pivot_table(index='year', columns='Violation', aggfunc=len, fill_value=0).reset_index()
lansing_yr_df['total'] = lansing_yr_df['Grass/Weed'] + lansing_yr_df['Trash']
lansing_yr_df.sort_values(by=['year'], inplace=True, ascending=False)
lansing_yr_df.head(10)
```

```python
#***Violations - year breakdown***
lansing_df['year'] = pd.DatetimeIndex(lansing_df['Date Filed']).year
print(lansing_df.head(10))
lansing_yr_violation_df = lansing_df.groupby(['year','Violation'])['Violator Address'].count().reset_index(name='Violations Count')
lansing_yr_violation_df.head(10)
```

```python
alt.Chart(lansing_yr_violation_df).mark_bar().encode(
    column=alt.Column('year', spacing=30, sort='descending', header=alt.Header(title='', labelOrient='bottom')),
    x=alt.X('Violation', axis=None),
    y='Violations Count:Q',
    color=alt.Color('Violation:N',scale=alt.Scale(domain=['Grass/Weed', 'Trash'],range=['green', 'black'])),
).configure_axis(
    grid=False
).configure_view(
    strokeOpacity=0
).properties(width=50, height=300)
```

```python
# Add 'Season' column
# Ref:https://stackoverflow.com/questions/44526662/group-data-by-season-according-to-the-exact-dates
bins = [0, 91, 183, 275, 366]
labels=['Winter', 'Spring', 'Summer', 'Fall']
doy = lansing_df['Date Filed'].dt.dayofyear
lansing_df['Season'] = pd.cut(doy + 11 - 366*(doy > 355), bins=bins, labels=labels)
lansing_yr_season_df = lansing_df.groupby(['year','Violation','Season'])['Violator Address'].count().reset_index(name='Violations Count')
# view df after adding new column for Season
lansing_yr_season_df.head(30)
```

```python
lansing_yr_season_grass_df = lansing_yr_season_df[lansing_yr_season_df.Violation=='Grass/Weed']
seasons_sort=['Spring','Summer','Fall','Winter']
alt.Chart(lansing_yr_season_grass_df).mark_bar().encode(
    column=alt.Column('year', spacing=30,sort='descending',header=alt.Header(title='', labelOrient='bottom')),
    x=alt.X('Season', sort=seasons_sort, axis=None),
    y='Violations Count:Q',
    color=alt.Color('Season:N',scale=alt.Scale(domain=['Spring', 'Summer','Fall','Winter'],range=['green', 'yellow','orange','#add1eb']))
).configure_axis(
    grid=False
).configure_view(
    strokeOpacity=0
).properties(width=100, height=350)
```

```python
lansing_yr_season_trash_df = lansing_yr_season_df[lansing_yr_season_df.Violation=='Trash']
alt.Chart(lansing_yr_season_trash_df).mark_bar().encode(
    column=alt.Column('year', spacing=30,sort='descending',header=alt.Header(title='', labelOrient='bottom')),
    x=alt.X('Season', axis=None, sort=['Spring', 'Summer','Fall','Winter']),
    y=alt.Y('Violations Count:Q'),
    color=alt.Color('Season:N',sort=alt.SortField(field="y", order='descending') ,scale=alt.Scale(domain=['Spring', 'Summer','Fall','Winter'],range=['green', 'yellow','orange','#add1eb']))
).configure_axis(
    grid=False
).configure_view(
    strokeOpacity=0
).properties(width=100, height=350)
```
